```python
import yfinance as yf
import pandas as pd
import time
import datetime
from datetime import datetime, timedelta
```

This is for all US Stocks Only (Excludiing ETF/Bonds)

This is S&P 500

```python
stock_symbol = pd.read_csv("S&P500-Symbols.csv") ##File to read coressponding to name
```

```python
stock_symbol
```

```python
list_of_tickers_sp500 = stock_symbol['Symbol'].tolist() 
```

```python
list_of_tickers_sp500.append('^IXIC')
list_of_tickers_sp500.append('^gspc')
list_of_tickers_sp500.append('^DJI')
list_of_tickers_sp500.append('^RUT')
list_of_tickers_sp500.append('CL=F')
list_of_tickers_sp500.append('GC=F')
list_of_tickers_sp500.append('^RUT')
```

```python
list_of_tickers_sp500
```

```python
for i in list_of_tickers_sp500:
    ticker=yf.Ticker(i)
    df=ticker.history(period="1d")
    current_time = datetime.now()
    daily_date = current_time.strftime('%m_%d_%Y')
    pd_daily = pd.DataFrame(df)
    name='Ticker_Values_market.csv'
    pd_daily['Ticker']=i
    pd_daily.to_csv(name,mode='a', header=False)
    #print(df)
    
```
