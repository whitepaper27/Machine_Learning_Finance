```python
my_first_tickers=['AAPL']
for i in my_first_tickers:
    ticker=yf.Ticker(i)
    df=ticker.history(period="ytd")
    current_time = datetime.now()
    daily_date = current_time.strftime('%m_%d_%Y')
    pd_daily = pd.DataFrame(df)
    name='Stock'+'_'+daily_date + '.csv'
    pd_daily['Ticker']=i
    pd_daily.to_csv(name, index=True)
    #print(df)
```
