```python
stock_symbol = pd.read_csv("stock_sahil.csv") ##File to read coressponding to name
my_day=1#No. of days to gather data
apiKey='b3ef5c3d00a340a7b80af816ea263bc5-eDgV2ljVhYQ'#####Api Key hwew
```

```python
tod = datetime.datetime.now()
d = datetime.timedelta(days=my_day)
no_of_days = (tod - d).strftime("%Y-%m-%d")
```

```python
l_list=[]
MY_STATUS='NOT'
MY_RESULTS=0
for i, j in stock_symbol.iterrows(): 
    ticker=j['Symbol']
    #print(ticker)
    url = 'http://newsapi.org/v2/everything?'+'q=' + ticker+'&'+'from='+no_of_days+'&'+'sortBy=popularity&'+'apiKey='+apiKey
    response = requests.get(url)
    data =response.json()  
    my_keys = data.keys()
    for key in my_keys:
      #print('Key',key)
      all_values=data[key]
    #Status
      if key=='status':
        if all_values=='ok':
         MY_STATUS=all_values
#Total results
      if key=='totalResults':        
         MY_RESULTS=all_values        
      if key=='articles':
         all_values=data[key]
         for i in range(0,len(all_values)):
            #for j in all_values[i].keys():
                #print('In j',j)
                #print('content',all_values[i][j])
                 my_dict = {
                     "ticker":ticker,
                     "publishedAt": all_values[i]['publishedAt'],
                                           "content": all_values[i]['content'],
                                           "url": all_values[i]['url'],
                                           "description": all_values[i]['description'],
                                           "title": all_values[i]['title']
                                           }
                 l_list.append(my_dict)
```
